```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Read the dataset to python environment
assgn_data = pd.read_csv("diabetes.csv")
```

```python
# 2. Display the columns in the dataset.
assgn_data.columns
```

```python
assgn_data.info()
```

```python
# 3. Calculate the mean of each column of the dataset.
assgn_data[['Pregnancies','Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']].mean()
```

```python
# 4. Check for the null values present in the dataset.
assgn_data.isnull().sum()
```

```python
# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.
```

```python
# Scatter Plot of BMI vs. Age
plt.figure(figsize=(10,5))
plt.scatter(assgn_data['Age'], assgn_data['BMI'], s=20, c= 'red')
plt.xlabel("Age")
plt.ylabel("BMI")
plt.show()
```

```python
# Heat map
import seaborn as sns
assgn_data_new = assgn_data.drop(['Outcome'], axis =1)
assgn_data_new.corr()
```

```python
sns.heatmap(assgn_data_new.corr())
plt.xticks(rotation = 90)
```

```python
# Histogram of glucose levels
plt.figure(figsize=(8, 6))
plt.hist(assgn_data['Glucose'], bins=20, color='green')
plt.xlabel('Glucose Levels')
plt.ylabel('Frequency')
plt.title('Histogram of Glucose Levels')
plt.show()
```

```python
# boxplot of glucose level by pregancies
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pregnancies', y='Glucose', data=assgn_data, color = 'yellow')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Glucose Levels')
plt.title('Boxplot of Glucose Levels by Number of Pregnancies')
plt.show()
```
